```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/sdg_08_20_linear_2_0.csv')
```

```python
df.head()
```

```python
df = df[['Geopolitical entity (reporting)','OBS_VALUE']]
```

```python
df.head()
```

```python
df = df.rename(columns = {'Geopolitical entity (reporting)':'country','OBS_VALUE':'NEET_Rate'})
```

```python
df = df.dropna(subset=['country', 'NEET_Rate'])
```

```python
df = df.groupby('country')['NEET_Rate'].mean().reset_index()
```

```python
df = df.drop(labels=[9, 10], axis=0)
```

```python
df['NEET_Rate'] = df['NEET_Rate'].replace(0, pd.NA)
```

```python
df.to_csv('../clean_data/choropleth_data.csv', index=False)
```
